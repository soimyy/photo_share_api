```python
def make_divisors(n):
    divisors = []
    for i in range(1, int(n**0.5)+1):
        if n % i == 0:
            divisors.append(i)
            if i != n // i:
                divisors.append(n//i)

    # divisors.sort()
    return divisors
```

```python
divisors = make_divisors(1234567890)
```

```python
select_sum = [ num for num in divisors if num <= 5000000]
```
